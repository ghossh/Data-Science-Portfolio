```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('l1-492.csv')
```

```python
data
```

```python
x_speed=data.speed*np.cos(data.Direction)
x_speed
```

```python
y_speed=data.speed*np.sin(data.Direction)
y_speed
```

```python
plt.figure(figsize=(16, 8)) 
plt.subplot(2,1,1)
plt.title('Current Speed')

plt.plot(data.jday,x_speed,color='Red')
plt.ylabel('Zonal speed',fontsize='20')

plt.subplot(2,1,2)
plt.plot(data.jday,y_speed)
plt.ylabel('Meridional speed')
plt.xlabel('Julian Day')
```
